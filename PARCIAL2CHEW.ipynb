##### INCISO D

```python
%sql SELECT DISTINCT cliente.nombre AS cliente1, cust.nombre AS cliente2, cliente.credito FROM cliente INNER JOIN (SELECT * FROM cliente) AS cust ON cliente.credito = cust.credito WHERE cliente.nombre != cust.nombre;
```

##### INCISO E

```python
%sql SELECT pink.nombre FROM ( SELECT cliente.nombre, tienda.nombre FROM cliente INNER JOIN compra ON cliente.id = compra.cliente_id INNER JOIN tienda ON (tienda.id = compra.tienda_id) WHERE tienda.nombre = 'Meykos' ) AS pink INNER JOIN ( SELECT cliente.nombre, tienda.nombre FROM cliente INNER JOIN compra ON cliente.id = compra.cliente_id INNER JOIN tienda ON (tienda.id = compra.tienda_id) WHERE tienda.nombre != 'Le Club' ) AS encanto ON encanto.nombre = pink.nombre;
```

##### INCISO F

```python
%sql SELECT DISTINCT tienda.nombre AS tienda1, tienda2.nombre AS tienda2, tienda.direccion FROM tienda INNER JOIN (SELECT * FROM tienda) AS tienda2 ON tienda.direccion = tienda2.direccion WHERE tienda.nombre != tienda2.nombre;
```
